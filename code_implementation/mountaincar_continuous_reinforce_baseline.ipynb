```python
from train.actor_critic import REINFORCEwithBaseline, MountainContinuousActorV2, MountainContinuousCritic
import gymnasium as gym
import matplotlib.pyplot as plt
import torch
torch.autograd.set_detect_anomaly(True)
```

```python
env = gym.make("MountainCarContinuous-v0")
actor = MountainContinuousActorV2(hidden_dim=64)
critic = MountainContinuousCritic(hidden_dim=64)
trainer = REINFORCEwithBaseline(env=env, actor=actor, critic=critic)
```

```python
# Test training
episode_rewards = trainer.train(
    num_episodes=5,
    logging_per_episodes=1,
    save_per_episodes=1,
)

plt.figure(figsize=(8,4))
plt.plot(episode_rewards)
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("REINFORCE(baseline) Training Rewards")
plt.show()
```

```python
episode_rewards = trainer.train(
    num_episodes=500,
    logging_per_episodes=25,
    save_per_episodes=25,
)

plt.figure(figsize=(8,4))
plt.plot(episode_rewards)
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("DQN Training Rewards")
plt.show()
```
