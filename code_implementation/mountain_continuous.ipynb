```python
import torch
# torch.save(trainer.actor.state_dict(), "mountaincar_20250223/500.pth")
import gymnasium as gym
from train.actor_critic import REINFORCEBatch, MountainContinuousActor, REINFORCEBatchRenderer
import matplotlib.pyplot as plt
```

```python
actor = MountainContinuousActor(hidden_dim=16)
actor.load_state_dict(torch.load("mountaincar_20250223/500.pth"))
env = gym.make("MountainCarContinuous-v0", render_mode="rgb_array")
trainer = REINFORCEBatch(env=env, actor=actor, critic=None)
renderer = REINFORCEBatchRenderer(env, trainer)
frames_per_episode = renderer.rendering(max_step=1000)
```

```python
save_dir = "mountaincar_20250223/videos"
for i, frames in enumerate(frames_per_episode):
    save_name=f"500_{i}"
    renderer.save_as_video(frames, save_dir, save_name)
```

```python
# REINFORCEBatch
actor = MountainContinuousActor(hidden_dim=16)
trainer = REINFORCEBatch(env=env, actor=actor, critic=None)
trainer.train(
    beta=None,
    num_episodes=1000,
    logging_per_episodes=50,
)
```
